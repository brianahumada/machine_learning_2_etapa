#OTROS BORRADOR de CODIGO - Mis primeras NEURONAS



```python
##########################################################################3
#__________-LA CREACION de MIS PRIMERAS REDES NEURONALES___________________


###################################
#Data ANALITYC:
  #No codea tanto, pero si sabe algo de programacion
  #Sabe un poquito de matematica y un poquito de programacion
  #Habla con DEPARTAMENTOS y DIRECTIVOS para saber sobre el problemas
  # La informacion del DATA SCIENCE le sirve para verificar de todos los
  # datos que recopilo cuales son IMPORTANTES y PRIORITARIOS y cuales no

#Data SCIENCE:
  #Es un ingeniero o Licenciado o Profesional que aplica como un
  # MINERO DE DATOS, hace mineria a traves de el uso de MILES DE BASES DE DATOS
  # que la empresa le entrega o servidores para sacar informacion util,
  # segun la indicacion del ANALITYC y del ENGINNER.

#Data ENGINNER:
  #Crea MODELOS de ia de aprendizaje y a veces DIRIJE  a los
  # Data ANALITYCS y Data SCIENCE.
  # CREAR MODELOS QUE SON ALIMENTADOS POR bases de datos y data set etc,
  # que le provee el DATA SCIECNE, luego le da indicaciones al ANALITYC
  # para que pueda hacer los informes y sacar conclusiones.

#####################################################################################################################
#_____________________________________ALGORITMO CLASIFICACION_______________________________________________________
#####################################################################################################################

###################################
#_________PROCESO______________

###################################
#1-Importar LIBRERIAS:
import tensorflow as tf



###################################
#2- Crear un DATA SET de frutas()
# Imagina que tenemos frutas con dos características: tamaño y peso
frutas = tf.constant([[0.5, 0.2],  # Tamaño y peso de una fruta
                      [0.3, 0.15],
                      [0.2, 0.1],
                      [0.6, 0.25]])


###################################################
#3- Etiquetas de las frutas (0: Manzana, 1: Banana)
# Esto nos dice qué fruta es cada una es
etiquetas = tf.constant([0, 0, 1, 1])


###################################################################################3
#__________________________PRIMER RED NEURONAL_______________________________________

###################################################
#4- Crear la primer red neuronal
#Vamos a hacer una MAQUINA  que aprenda de las frutas:
#VEANLO como CAJAS FILTRADORAS y CAPAS (Capas de filtro y CAPAS FINAL)
modelo1= tf.keras.Sequential(
    [tf.keras.layers.Dense(units=4, activation='relu', input_shape=(2,)),  # Capa oculta con 4 "cajitas"
    tf.keras.layers.Dense(units=2, activation='softmax')  # Capa final con 2 "cajitas" para decir si es manzana o banana
     ])



###################################################
#5- Compilar el modelo
#Hacer que la "Maquina" este lista para aprender
modelo1.compile(optimizer="adam", loss='sparse_categorical_crossentropy', metrics['accuracy'])



###################################################
#6- Entrenar el MODELO1:
#Enseñarle a la "maquina" como clasificar frutas
modelo1.fit(frutas, etiquetas, epochs=50)

###################################################################################3
#__________________________SEGUNDA RED NEURONAL_______________________________________

###################################################
#Crear la SEGUNDA RED NEURONAL
#Ahora vamos a CREAR UNA "MAQUINA 2" distinta
modelo2 = tf.keras.Sequential([
    tf.keras.layers.Dense(units=3, activation='relu', input_shape=(2,)),  # Capa oculta con 3 "cajitas"
    tf.keras.layers.Dense(units=2, activation='softmax')  # Capa final con 2 "cajitas" para decir si es manzana o banana
])

###################################################
# Compilar el modelo2
# Preparar la nueva "máquina" para aprender
modelo2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])


###################################################
# Entrenar el modelo2
# Enseñarle a la nueva "máquina" cómo clasificar las frutas
modelo2.fit(frutas, etiquetas, epochs=50)


#######################################################################
#________________RESULTADOS DE MIS REDES NEURONALES_________________
#Evaluar LOS MODELOS
#Ver si nuestras "MAQUINAS" (Redes Neuronales) aprendieron a CLASIFICAR (perceptron)
#Algoritmo de PERCEPTRON: Define DOS CATEGORIAS
 #(Si es x o si es y(segun el data set si es una banana o es una manzana))

print("\nResultados del MODELO 1")
modelo1.evaluate(frutas, etiquetas)


print("\nResultados del MODELO 2")
modelo2.evaluate(frutas, etiquetas)













```
