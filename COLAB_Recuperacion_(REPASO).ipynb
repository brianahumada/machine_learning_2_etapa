##**(OPCIONAL)HERRAMIENTA 3: EJEMPLO de LANZAMIENTO de DADOS y NUMPY**


###EJEMPLO de LANZAMIENTO de DADOS y NUMPY

```python
#########################################################
#______MIS APUNTES - SEMANA 3 - PROBABILIDAD Y ESTADISTICA - Clase 1__________

#Probabilidad = posibilidad = Tirar los dados y sacar datos

###################################
#VENTAJAS de la probabilidad:

#     1- Predecir eventos futuros (basados en dato historicos)

###################################
#REPASO de DATA: Diferencia entre datas:
#   Recibo datos -> Limpio esos datos -> Proceso  -> A partir del procesamiento ->
#     -> Saco conclusiones

##################################################################
#PROCESO: (DADOS, papel y lapiz)

#     1- Lanzamiento
#     2- Guardar los registros de los lanzamientos
#     3- Hacer el calculos de PROBABILIDAD a esos registros

#     4- Graficar con una libreria de python (MATPLOTLIB)

####################################################################
#Experimento:

#     Dado regular: (PUNTAS PERFECTAS)Sus extremos estan bien definidos= forma casi perfecto y plana

#     Dadoo irregular:(PUNTAS REDONDAS) Sus extremos NO ESTAN bien definidos= forma imperfecta



####################################################################
#_______________CODIGO - Experimento de los dados en NUMPY__________________:

############################
#Importar la libreria:
import numpy as np

############################
#1- Lanzamientos: Generamos un vector que lo rellene con 100 numero aleatorios entre el 1 y 6 (se pone 7 porque el vector siempre va hasta el tope -1 es decir 7-1)
lanzamientos= np.random.randint(1, 7, 100) #()Genera numeros aleatorios ENTRE el 1 y el 6, CUANTOS= 100
print(lanzamientos)


################################################################
#REGISTROS:

#VECTOR= Puro y exclusivamente para GUARDAR los REGISTROS del 2
#[dado=2,    dado=-,    dado=-,    dado=2,    dado=2]
#   True     False     False      True      True
#    1          0         0         1          1


#VECTOR= Puro y exclusivamente para GUARDAR los REGISTROS de cuantas veces toco 3

#VECTOR= Puro y exclusivamente para GUARDAR los REGISTROS de cuantas veces toco 6

#VECTOR= Puro y exclusivamente para GUARDAR los REGISTROS de cuantas veces toco 1


#######################################################
#2- Crear REGISTRO:
#(escribir en papel mis resultados) - Crear vector= verdaero, falso, verdader...etc
#Vector BOOLEANS= Booleanos= 2 valores= VERDADERO(True), FALSO(False)= Prendido 1, Apagado 0

#_______Crear registro= VECTOR________
#Vector que indica cuantas veces dio 2
prob2_bool= lanzamientos == 2 #lanzamiento= [4,2,6,7,2] --> prob1_bool= [false, true, false, false, true]
prob3_bool= lanzamientos == 3

#ALTERNATIVA: Crear una variable "contador_2",
#             crear un for, con el for recorrer "lanzamientos", y mientras for recorre el vector
#             lanzamiento, contador cuenta las veces que toco.


################################
#4-Crear CONTADOR para sumar todas las veces que dio TRUE, es decir todas las veces que salio 1
prob2_contador= np.sum(prob2_bool)
prob3_contador= np.sum(prob3_bool)


###############################
#3- Aplicar CALCULO de PROBABILIDAD
# En una variabla "prob1" guardar= contador/total:

prob2= prob2_contador/ len(lanzamientos) #len: Es una funcion para saber la longitud de un vector
prob3= prob3_contador/ len(lanzamientos)


###############################
#4- MOSTRAR Resultados:
print(f"probabilidad de obtener 2: {prob2: .2f}") #variable: .2f =este codigo es para mostrar una cant(2) decimales
print(f"probabilidad de obtener 3: {prob3: .2f}")


###############################
#5- GRAFICAMOS= HISTOGRAMA (Un diagrama que cuenta una historia, mostramos la probabilidad):

#importar libreria
import matplotlib.pyplot as plt

numeros= [2,3] # En un vector poner lo que tenemos que graficar

probabilidades= [prob2, prob3] #En un vector probabilidades el resultado prob2

#INGLES:
#   bar: BARRA=   plt.bar (numeros, probabilidaes)
#   label: NOMBRE(etiqueta)= EJES=   plt.xlabel, plt.ylabel
#   title: TITULO= plt.title("MI GRAFIQUITO, o HISTOGRAMA DE PROBABILIDADES")
#   show:  MOSTRAR= plt.show()

plt.bar(numeros, probabilidades)

plt.xlabel('numeros')
plt.ylabel('probabilidad')
plt.title('MI GRAFIQUITO, o HISTOGRAMA DE PROBABILIDADES')

plt.show()


```
